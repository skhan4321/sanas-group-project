#### Data Cleaning

```python
# The data set is mostly clean but has a lot of null values

df_backup = df_backup.dropna(how='any')
df_backup
```

```python
zip_df = df_backup["Zip Code"].value_counts().sort_values(ascending = False)
zip_df
```

```python
zip_df.plot(kind = 'bar', color = 'red', figsize = (12,8))
plt.xticks(rotation=60)
plt.xlabel('Zip Codes')
plt.ylabel('Number of Calls')
plt.show()
```

---
#### Data Study Filtered by Shared Micromobility 

```python
shared_mobility_calls = df_backup[df_backup["SR Description"] == "Shared Micromobility"]
shared_mobility_calls
```

```python
sm_zip = shared_mobility_calls["Zip Code"].value_counts()
sm_zip
```

```python
sm_mobi = (df_backup[df_backup["SR Type Code"]== "DOCKMOBI"])&&(df_backup[df_backup["SR Description"]== "Shared Micromobility"])
sm_mobi
```

```python
# Plot the number of calls related to shared micromobility  by zipcode:
sm_zip.plot(kind = 'bar', color = 'red', figsize = (12,8))
plt.xticks(rotation=60)
plt.xlabel('Zip Codes')
plt.ylabel('Number of Calls')
plt.show()
```

#### Research Result:
Clearly shows that most number of calls regarding shared micromobility were from zipcodes 78701 and 78704.
